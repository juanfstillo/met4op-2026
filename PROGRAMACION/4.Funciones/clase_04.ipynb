## Estructura

Una función se define usando la palabra clave de bloque "def" , seguida del nombre de la función, seguido de un par de paréntesis que contienen los **parámetros** que tomará la función y que terminan con dos puntos.  
A continuación, sigue el **bloque de declaraciones** que forman parte de esta función.   
Las funciones pueden devolver un valor a la persona que llama, utilizando la palabra clave 'retorno'.  

### Estructura para Definir una Funciones:

```python
def fn(par1, par2,...):
	"""docstring"""
	CODIGO
	return
```

Un parámetro es la variable que forma parte de la firma del método (declaración del método). Los parámetros se especifican dentro del par de paréntesis en la definición de la función, separados por comas. Cuando llamamos a la función, proporcionamos los valores reales para dichos parametros, y nos referiremos a ellos como **Argumentos**.



### Llamar a una Función

### Parámetros Mutables o Inmutables

Los argumentos para la funciones siempre se pasan como una referencia a a un espacio de memoria, y luego se crea una variable local que comparte dicho espacio de memoria. Ahora bien, si modificamos el argumento dentro de la función se obtendran comportamientos distintos dependiendo de si el objeto es mutable o no 

* integer, float, string y tuple son objetos inmutables

* list, dict y set  son objetos mutables

Esto implica que el estado de los segundos pueden modificarse por la función

### Variables Locales vs Globales

En Python, el lugar donde creas una variable determina desde dónde puedes usarla (su scope o alcance). Las variables globales viven en todo el programa, mientras que las locales nacen y mueren exclusivamente dentro de la función donde fueron creadas.

````python
mensaje_global = "Soy una variable global"   # VARIABLE GLOBAL: Declarada fuera de cualquier función.  # Todo el programa (incluso las funciones) puede "verla".

def mi_funcion(parametro):
    """Demostración de scope local y global."""
    
    mensaje_local = "Soy una variable local"
    print("estoy dentro de la función")
    print(mensaje_global)  
    print(mensaje_local)  # VARIABLE LOCAL: Declarada dentro de la función.  # Solo existe mientras la función se está ejecutando.

mi_funcion("test")
print("estoy fuera de la función")
print(mensaje_global)
print(mensaje_local)  #<-- Si intentas hacer esto afuera, Python dará un error (NameError), 
````

## Funciones Recursivas

Una función **recursiva** es aquella que se llama a sí misma dentro de su propio cuerpo. Para que no se ejecute infinitamente, toda función recursiva necesita:

1. Un **caso base**: la condición que detiene la recursión.
2. Un **caso recursivo**: donde la función se llama a sí misma con un problema más pequeño.

La recursión es especialmente útil cuando un problema puede dividirse en versiones más pequeñas del mismo problema.

### Ejemplo clásico: Factorial

El factorial de n (escrito n!) se define como:
- `0! = 1` (caso base)
- `n! = n × (n-1)!` (caso recursivo)

Esto es naturalmente recursivo: el factorial de n depende del factorial de n-1.

````python
def factorial(n):
    """Calcula el factorial de n de forma recursiva."""
    # Caso base: el factorial de 0 es 1
    if n == 0:
        return 1
    # Caso recursivo: n! = n * (n-1)!
    else:
        return n * factorial(n - 1)

# ¿Cuántas formas distintas hay de ordenar 5 candidatos en una boleta?
print("Formas de ordenar 5 candidatos:", factorial(5))
print("Formas de ordenar 3 candidatos:", factorial(3))
````

### ¿Cómo funciona internamente?

Cuando llamamos `factorial(4)`, Python apila las llamadas así:
```
factorial(4)
  → 4 * factorial(3)
       → 3 * factorial(2)
            → 2 * factorial(1)
                 → 1 * factorial(0)
                      → 1   ← caso base
```
Luego va resolviendo hacia atrás: 1×1=1, 2×1=2, 3×2=6, 4×6=**24**

### Ejemplo aplicado: Suma acumulada de votos

Imaginen que queremos sumar los votos de una lista de manera recursiva (equivalente a `sum()`, pero para entender la recursión):

````python
def suma_votos(lista_votos):
    """Suma una lista de votos de forma recursiva."""
    # Caso base: lista vacía, suma = 0
    if len(lista_votos) == 0:
        return 0
    # Caso recursivo: primer elemento + suma del resto
    else:
        return lista_votos[0] + suma_votos(lista_votos[1:])

votos_por_provincia = [234567, 189023, 412890, 98765, 321456]
total = suma_votos(votos_por_provincia)
print("Total de votos:", total)
print("Verificación con sum():", sum(votos_por_provincia))
````

### Importante: el límite de recursión en Python

Python tiene un límite de profundidad de recursión (~1000 llamadas). Si una función se llama a sí misma demasiadas veces, Python lanzará un `RecursionError`. Por eso, para grandes volúmenes de datos es preferible usar bucles o la función `sum()` de Python.

## Funciones de Orden Superior

Una **función de orden superior** es aquella que recibe otras funciones como argumentos o las devuelve como resultado. Esto permite escribir código más general y reutilizable.

Las tres más importantes en Python son `map()`, `filter()` y `sorted()`.

### `map()`: aplicar una función a cada elemento

`map(funcion, iterable)` aplica `funcion` a cada elemento del iterable y devuelve un nuevo iterador con los resultados.

````python
partidos = ['frente de todos', 'juntos por el cambio', 'consenso federal']

# Poner en mayúsculas todos los nombres
partidos_mayus = list(map(lambda p: p.upper(), partidos))
print(partidos_mayus)

# Calcular el porcentaje de votos para cada partido
votos = [4890523, 4028853, 1603038]
total_votos = sum(votos)

porcentajes = list(map(lambda v: round(v / total_votos * 100, 2), votos))
print(porcentajes)
````

### `filter()`: filtrar elementos según una condición

`filter(funcion, iterable)` devuelve solo los elementos para los que la función retorna `True`.

````python
# Filtrar partidos que superaron el umbral del 1.5% (piso electoral)
umbral = 0.015
votos_partidos = {
    'FRENTE DE TODOS': 4890523,
    'JUNTOS POR EL CAMBIO': 4028853,
    'CONSENSO FEDERAL': 1603038,
    'FRENTE DE IZQUIERDA': 760154,
    'MOVIMIENTO AL SOCIALISMO': 34891,
    'PARTIDO AUTONOMISTA': 2341,
}
total = sum(votos_partidos.values())

# Filtrar: solo los que superan el piso
supera_piso = list(filter(
    lambda partido: votos_partidos[partido] / total >= umbral,
    votos_partidos
))
print("Partidos que superaron el piso electoral:")
for p in supera_piso:
    pct = round(votos_partidos[p] / total * 100, 2)
    print(f"  {p}: {pct}%")
````

### `sorted()` con clave personalizada

`sorted(iterable, key=funcion)` ordena elementos usando la función como criterio.

````python
partidos_votos = [
    ('FRENTE DE TODOS', 4890523),
    ('JUNTOS POR EL CAMBIO', 4028853),
    ('CONSENSO FEDERAL', 1603038),
    ('FRENTE DE IZQUIERDA', 760154),
    ('UNITE', 134380),
]

# Ordenar de mayor a menor cantidad de votos
ranking = sorted(partidos_votos, key=lambda x: x[1], reverse=True)

print("Ranking electoral 2019:")
for i, (partido, votos) in enumerate(ranking, 1):
    print(f"  {i}. {partido}: {votos:,} votos")
````

### `.apply()` sobre columnas: crear nuevas variables

`.apply()` aplica una función a cada elemento de una Serie (columna) o a cada fila de un DataFrame. Es la herramienta más flexible para transformar datos.

````python
# Función para clasificar el nivel de votos de una agrupación en una mesa
def clasificar_votacion(votos):
    """Clasifica la cantidad de votos por mesa en categorías."""
    if votos == 0:
        return 'Sin votos'
    elif votos <= 10:
        return 'Baja'
    elif votos <= 50:
        return 'Media'
    else:
        return 'Alta'

# Aplicamos la función a la columna VOTOS_AGRUPACION
votos_partidos['nivel_votacion'] = votos_partidos['VOTOS_AGRUPACION'].apply(clasificar_votacion)

# Veamos la distribución
print(votos_partidos['nivel_votacion'].value_counts())
````

````python
# Lo mismo con lambda (versión compacta)
votos_partidos['nivel_lambda'] = votos_partidos['VOTOS_AGRUPACION'].apply(
    lambda v: 'Sin votos' if v == 0 else ('Baja' if v <= 10 else ('Media' if v <= 50 else 'Alta'))
)

# Verificamos que son iguales
print("¿Son iguales?", (votos_partidos['nivel_votacion'] == votos_partidos['nivel_lambda']).all())
````

### Funciones que trabajan sobre grupos: `groupby()` + funciones personalizadas

Una de las tareas más comunes en análisis electoral es agregar datos por partido, región o distrito. Podemos combinar `groupby()` con funciones propias.

````python
# Función que calcula el resumen de votos para un grupo
def resumen_votos(serie):
    """Devuelve estadísticas básicas de votos para un grupo."""
    return pd.Series({
        'total_votos': serie.sum(),
        'promedio_por_mesa': round(serie.mean(), 2),
        'max_en_mesa': serie.max(),
        'mesas_sin_votos': (serie == 0).sum()
    })

# Aplicamos por agrupación política
resumen_por_partido = (
    # Seleccionamos la columna correctamente y la comparamos por fuera del corchete interno
    votos_partidos[votos_partidos['CODIGO_CATEGORIA'] == 100000000000] 
    .groupby('NOMBRE_AGRUPACION')['VOTOS_AGRUPACION']
    .apply(resumen_votos)
    .unstack()
    .sort_values('total_votos', ascending=False)
)

print("Resumen de votos por partido:")
print(resumen_por_partido.head(10))
````

### Funciones que calculan porcentajes dentro de grupos

En análisis electoral, necesitamos frecuentemente calcular el porcentaje de votos dentro de cada categoría geográfica. Para eso usamos `transform()`.

````python
principales = ['FRENTE DE TODOS', 'JUNTOS POR EL CAMBIO', 'CONSENSO FEDERAL']

# Todo el cálculo y filtrado en un solo flujo (inline)
muestra_region = (
    votos_partidos[votos_partidos['CODIGO_CATEGORIA'] == 100000000000]
    .groupby(['NOMBRE_REGION', 'NOMBRE_AGRUPACION'], as_index=False)['VOTOS_AGRUPACION'].sum()
    .assign(
        pct_en_region=lambda x: round(x['VOTOS_AGRUPACION'] / x.groupby('NOMBRE_REGION')['VOTOS_AGRUPACION'].transform('sum') * 100, 2)
    )
    .query('NOMBRE_AGRUPACION in @principales')
    .sort_values(['NOMBRE_REGION', 'VOTOS_AGRUPACION'], ascending=[True, False])
)

print("Porcentaje de votos a Presidente por Región (Partidos Principales):")
print(muestra_region.head(15).to_string(index=False))
````

### `.pipe()`: encadenamiento limpio de funciones

`.pipe()` permite encadenar funciones de manera legible, como si fuera una 'cadena de procesamiento' del DataFrame. Es muy útil para pipelines de análisis de datos.

````python
def filtrar_categoria(df, codigo_categoria):
    """Filtra el dataset por una categoría electoral específica."""
    # 100000000000 es el código para Presidente
    return df[df['CODIGO_CATEGORIA'] == codigo_categoria]

def filtrar_partidos_principales(df, min_votos=100000):
    """Filtra los partidos con más de min_votos en total."""
    totales = df.groupby('NOMBRE_AGRUPACION')['VOTOS_AGRUPACION'].sum()
    partidos_grandes = totales[totales >= min_votos].index
    return df[df['NOMBRE_AGRUPACION'].isin(partidos_grandes)]

def agregar_por_partido(df):
    """Agrupa por partido y suma los votos."""
    return df.groupby('NOMBRE_AGRUPACION')['VOTOS_AGRUPACION'].sum().reset_index()

def calcular_porcentaje_total(df):
    """Agrega una columna con el porcentaje del total."""
    df = df.copy()
    df['porcentaje'] = round(df['VOTOS_AGRUPACION'] / df['VOTOS_AGRUPACION'].sum() * 100, 2)
    return df.sort_values('porcentaje', ascending=False)

# Pipeline de análisis: encadenamos las funciones con .pipe()
resultado_final = (
    votos_partidos
    .pipe(filtrar_categoria, codigo_categoria=100000000000)  
    .pipe(filtrar_partidos_principales, min_votos=50000)
    .pipe(agregar_por_partido)
    .pipe(calcular_porcentaje_total)
)

print("Resultados electorales a Presidente 2019 (Partidos Principales):")
print(resultado_final.to_string(index=False))

print(f"Pocentaje de votos a principales partidos sobre total de votos: ",
round(resultado_final['VOTOS_AGRUPACION'].sum() / votos_partidos[votos_partidos['CODIGO_CATEGORIA'] == 100000000000]['VOTOS_AGRUPACION'].sum(), 2))
````

## Resumen: ¿Cuándo usar cada tipo de función?

| Tipo | Cuándo usarla |
|------|---------------|
| `def` estándar | Lógica compleja, reutilizable, con documentación |
| `lambda` | Transformaciones simples inline, con `map`/`filter`/`sorted` |
| Recursiva | Problemas que se dividen en versiones más pequeñas de sí mismos |
| Con `.apply()` | Aplicar lógica fila a fila o elemento a elemento en un DataFrame |
| Con `.groupby()` | Calcular estadísticas por grupos (partido, distrito, mesa) |
| Con `.pipe()` | Encadenar varios pasos de procesamiento de forma legible |
