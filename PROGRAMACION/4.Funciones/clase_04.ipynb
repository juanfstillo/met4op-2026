# FUNCIONES

## Estructura

Una función se define usando la palabra clave de bloque "def" , seguida del nombre de la función, seguido de un par de paréntesis que contienen los **parámetros** que tomará la función y que terminan con dos puntos.  
A continuación, sigue el **bloque de declaraciones** que forman parte de esta función.   
Las funciones pueden devolver un valor a la persona que llama, utilizando la palabra clave 'retorno'.  

### Estructura para Definir una Funciones:

```python
def fn(par1, par2,...):
	"""docstring"""
	CODIGO
	return
```

Un parámetro es la variable que forma parte de la firma del método (declaración del método). Los parámetros se especifican dentro del par de paréntesis en la definición de la función, separados por comas. Cuando llamamos a la función, proporcionamos los valores reales para dichos parametros, y nos referiremos a ellos como **Argumentos**.



### Llamar a una Función
```python
fn(arg1, arg2,...)
```

### Ejemplo

````python
def suma(x,y):
    """Mi fn"""
    out = x+y
    return out
````

````python
suma(1,2)
````

### Parámetros Mutables o Inmutables

Los argumentos para la funciones siempre se pasan como una referencia a a un espacio de memoria, y luego se crea una variable local que comparte dicho espacio de memoria. Ahora bien, si modificamos el argumento dentro de la función se obtendran comportamientos distintos dependiendo de si el objeto es mutable o no 

* integer, float, string y tuple son objetos inmutables

* list, dict y set  son objetos mutables

Esto implica que el estado de los segundos pueden modificarse por la función

#### Argumento Immutable

### Variables Locales vs Globales

En Python, el lugar donde creas una variable determina desde dónde puedes usarla (su scope o alcance). Las variables globales viven en todo el programa, mientras que las locales nacen y mueren exclusivamente dentro de la función donde fueron creadas.

````python
mensaje_global = "Soy una variable global"   # VARIABLE GLOBAL: Declarada fuera de cualquier función.  # Todo el programa (incluso las funciones) puede "verla".

def mi_funcion(parametro):
    """Demostración de scope local y global."""
    
    mensaje_local = "Soy una variable local"
    print("estoy dentro de la función")
    print(mensaje_global)  
    print(mensaje_local)  # VARIABLE LOCAL: Declarada dentro de la función.  # Solo existe mientras la función se está ejecutando.

mi_funcion("test")
print("estoy fuera de la función")
print(mensaje_global)
print(mensaje_local)  #<-- Si intentas hacer esto afuera, Python dará un error (NameError), 
````

### Argumentos con Valores por defecto

En Python podemos proporcionar valores predeterminados para los parámetros de función en Python en caso de que el usuario no desee proporcionar valores para ellos. Esto se hace con la ayuda de los valores de argumento predeterminados. El valor predeterminado se asigna mediante el operador de asignación (=).

````python
def ponderar_respuesta(respuesta, peso, factor_ajuste=1.2):
    resultado = respuesta * peso * factor_ajuste
    return resultado
respuesta = 1
peso = 0.8
ponderar_respuesta(respuesta, peso)
````

La principal ventaja del argumento predeterminado es que podemos dar valores solo a aquellos parámetros a los que queramos, siempre que los otros parámetros tengan valores de argumento predeterminados.

````python
ponderar_respuesta(respuesta,peso,factor_ajuste=1.7)
````

### Número variable de argumentos

A veces, los programas pueden querer definir una función que pueda tomar cualquier número de parámetros, es decir, un número variable de argumentos, esto se puede lograr usando las estrellas (*). Esto es muy útil cuando no sabemos el número exacto de argumentos que se pasarán a una función

````python
import math
def dispSquare(*varArgs):
    #Iterar sobre todos los valores 
    for num in varArgs:
        print("La  raiz cuadrada de",num,  "es :" , math.sqrt(num) )

dispSquare(1,2,3,4,5,6,7)
````

## Ejemplo

Supongamos que hacemos una pregunta sobre la opinió que los encuestados tienen respecto a un candidato en una escal de 0 a 10

````python
Opinion = [10,1,0,7,3,5,2,7,5,8,9,2,4,6,9,1,2,8]
````

Primero vamos a construir una función que ordena los elementos. Para ello vamos a utilizar a modo de ejemplo el algoritmo Bubble-Sort

![](https://raw.githubusercontent.com/MET4OP-UBA/MET4OP/be8da8567334cbdb85fe6054721edeb50be99e40/clases/clase_03/extra/buble_sort.png)

````python

def bubbleSort(arr): 
    n = len(arr) 
  
     # Recorre todos los elementos de la matriz
    for i in range(n): 
        print("iteración", i, ":", arr)
         # Los últimos elementos i ya están en su lugar
        for j in range(0, n-i-1):   
            # atraviesa la matriz de 0 a n-i-1
             # Intercambiar si el elemento encontrado es mayor
             # que el siguiente elemento 
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j]

bubbleSort(Opinion)
````

````python


# Una versión optimizada de Bubble Sort
def bubbleSort(arr): 
    n = len(arr) 
   
    # Recorre todos los elementos de la matriz
    for i in range(n): 
        swapped = False
        print("iteración", i, ":", arr)
        # Los últimos i elementos ya están
        #  en su lugar 
        for j in range(0, n-i-1): 
   
            # atraviesa la matriz de 0 a
            # n-i-1. Cambiar si el elemento
            # encontrado es mayor que el
            # elemento siguiente 
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
                swapped = True
  
        # SI no se intercambiaron dos elementos
        # por bucle interno, luego romper 
        if swapped == False: 
            break

bubbleSort(Opinion)
````

Algunas características del ordenamiento tipo de burbuja:
* Los valores grandes siempre se ordenan primero.
* Solo se necesita una iteración para detectar que una colección ya está ordenada.
* La mejor complejidad de tiempo para Bubble Sort es O(n). La complejidad de tiempo promedio y peor es O(n²).
* La complejidad del espacio para Bubble Sort es O(1), porque solo se requiere un espacio de memoria adicional.

### Percentiles

````python

def bubbleSort(arr): 
    n = len(arr) 
    # Recorre todos los elementos de la matriz
    for i in range(n): 
        swapped = False
        # Los últimos i elementos ya están en su lugar 
        for j in range(0, n-i-1): 
            # atraviesa la matriz de 0 a n-i-1. Cambiar si el elemento
            # encontrado es mayor que el elemento siguiente 
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
                swapped = True
        # SI no se intercambiaron dos elementos
        # por bucle interno, luego romper 
        if swapped == False: 
            break

def resumen(arr):
    # Copio el arr, depositandolo en x
    x = arr.copy()
    # Calculo su tamaño
    n = len(x)
    # Ordeno en en lugar a x
    bubbleSort(x)
    # Calculo las medidas resumen
    minimo = x[0]
    maximo = x[-1]
    mediana = (x[math.floor((n+1)/2)-1] + x[math.ceil((n+1)/2)-1]) / 2.0
    Q1 = (x[math.floor((n+1)/4)-1] + x[math.ceil((n+1)/4)-1]) / 2.0
    Q3 = (x[math.floor(3*(n+1)/4)-1] + x[math.ceil(3*(n+1)/4)-1]) / 2.0

    return [minimo,Q1,mediana,Q3,maximo]

resumen(Opinion)
````
