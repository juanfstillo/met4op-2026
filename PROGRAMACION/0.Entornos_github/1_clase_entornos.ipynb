## ¿Por qué necesitamos un entorno de trabajo?

Antes de escribir una línea de código, necesitamos un lugar donde hacerlo. En Python esto no es trivial: distintos proyectos pueden requerir distintas versiones de librerías, y mezclarlas en una sola instalación genera conflictos.

La solución es trabajar con **entornos virtuales**: instalaciones aisladas de Python donde cada proyecto tiene sus propias dependencias.

En esta clase vemos tres formas de trabajar:

| Opción | Descripción | Ideal para |
|---|---|---|
| **Anaconda + VS Code** | Instalación local completa | Trabajo profesional y académico |
| **VS Code standalone** | Instalación mínima sin Anaconda | Usuarios más avanzados |
| **Google Colab** | Todo en el navegador, sin instalar nada | Arrancar rápido o trabajar desde cualquier lugar |

## 🌐 Entornos Virtuales con conda

Un **entorno virtual** es una instalación aislada de Python. Podés tener varios entornos en tu computadora, cada uno con sus propias librerías y versiones, sin que se interfieran entre sí.

```
Tu computadora
├── entorno: met4op       → Python 3.10, pandas 2.0, geopandas 0.14
├── entorno: otro_proyecto → Python 3.8, pandas 1.5
└── entorno: base          → instalación por defecto de Anaconda
```

### Comandos para gestionar entornos

````python
# === CREAR Y ACTIVAR ENTORNOS ===

# Crear un entorno nuevo con una versión de Python específica
# conda create -n met4op python=3.10

# Activar el entorno
# conda activate met4op

# Desactivar el entorno (volver al base)
# conda deactivate

# Ver todos los entornos disponibles
# conda env list
````

````python
# === INSTALAR Y GESTIONAR PAQUETES ===

# Instalar un paquete en el entorno activo
# conda install pandas

# Instalar desde pip (cuando conda no lo tiene)
# pip install geopandas

# Ver todos los paquetes instalados en el entorno
# conda list

# Actualizar un paquete
# conda update pandas

# Exportar el entorno (para compartirlo con otros)
# conda env export > environment.yml

# Recrear un entorno desde un archivo
# conda env create -f environment.yml
````

**VS Code** es el editor de código más usado en el mundo. Es gratuito, liviano y muy extensible. Funciona perfecto con Python y Jupyter Notebooks, y se integra directamente con Anaconda.

## 🔗 Conectar VS Code con Anaconda

Una vez instalados ambos, hay que decirle a VS Code qué intérprete de Python usar.

**Pasos:**

1. Abrir VS Code
2. Presionar `Ctrl+Shift+P` para abrir la paleta de comandos
3. Escribir **`Python: Select Interpreter`** y seleccionarlo
4. Elegir el entorno de conda que creaste (por ejemplo, `met4op`)

A partir de ahí, cada vez que ejecutes código o abras un notebook, VS Code va a usar las librerías de ese entorno.

> 💡 En la barra inferior de VS Code siempre podés ver qué intérprete está activo y cambiarlo con un click.

### Abrir un Jupyter Notebook en VS Code

1. Ir a `Archivo → Abrir archivo` y seleccionar un archivo `.ipynb`
2. O crear uno nuevo: `Ctrl+Shift+P` → **`Create: New Jupyter Notebook`**
3. En la esquina superior derecha del notebook aparece el kernel activo — verificar que sea el entorno correcto

VS Code permite ejecutar celdas con `Shift+Enter`, igual que Jupyter clásico.

## 🔧 VS Code sin Anaconda (instalación standalone)

Es posible usar VS Code con Python sin instalar Anaconda. Esta opción es más liviana pero requiere gestionar las dependencias manualmente con `pip` y el módulo `venv` de Python.

**Pasos:**

1. Instalar Python desde [https://python.org/downloads](https://python.org/downloads) (**marcar la opción "Add Python to PATH"** durante la instalación)
2. Instalar VS Code
3. Crear un entorno virtual desde la terminal:

````python
# === ENTORNO VIRTUAL CON venv (sin Anaconda) ===

# Crear un entorno virtual en la carpeta del proyecto
# python -m venv venv

# Activar el entorno
# En Windows:
# venv\Scripts\activate

# En macOS / Linux:
# source venv/bin/activate

# Instalar paquetes con pip
# pip install pandas numpy matplotlib seaborn jupyter

# Guardar las dependencias del proyecto
# pip freeze > requirements.txt

# Instalar dependencias desde un archivo
# pip install -r requirements.txt
````

### conda vs venv: ¿cuál usar?

| | `conda` (Anaconda) | `venv` (Python puro) |
|---|---|---|
| Instalación | Anaconda | Solo Python |
| Gestión de paquetes | `conda` + `pip` | Solo `pip` |
| Librerías científicas | Muy fácil (resuelve dependencias) | A veces complejo |
| Recomendado para | Principiantes y ciencia de datos | Usuarios avanzados / producción |

> Para esta materia, recomendamos **Anaconda**. Si ya tenés experiencia con Python y preferís una instalación más liviana, `venv` es perfectamente válido.

## ☁️ Google Colab

## 💡 Buenas prácticas

- **Usá un entorno por proyecto**: no instalés todo en el entorno `base` de Anaconda
- **Exportá tu entorno** con `conda env export` o `pip freeze` para que otros puedan reproducir tu trabajo
- **Agregá `venv/` y `.ipynb_checkpoints/` al `.gitignore`** para no subir basura al repositorio
- **En Colab, guardá una copia en Drive** antes de cerrar la sesión
- **Nombrá tus notebooks con números y descripción** (`01_limpieza.ipynb`, `02_analisis.ipynb`)
