## ¿Qué es Git?

**Git** es un sistema de control de versiones. Permite registrar los cambios que hacés sobre archivos a lo largo del tiempo, volver a versiones anteriores, y trabajar en equipo sin pisarse el trabajo.

Pensalo como un **historial de guardados** de tu proyecto, pero mucho más poderoso que Ctrl+Z.

## ¿Qué es GitHub?

**GitHub** es una plataforma web que usa Git por debajo. Te permite alojar tus repositorios en la nube, compartirlos, y colaborar con otras personas.

| | Git | GitHub |
|---|---|---|
| ¿Qué es? | Software local | Plataforma web |
| ¿Dónde vive? | En tu computadora | En internet |
| ¿Para qué? | Controlar versiones | Compartir y colaborar |

## 📁 Conceptos clave

| Concepto | Descripción |
|---|---|
| **Repositorio (repo)** | Carpeta de tu proyecto bajo control de versiones |
| **Commit** | Una foto del estado de tu proyecto en un momento |
| **Branch (rama)** | Una línea de trabajo paralela e independiente |
| **Merge** | Unir dos ramas |
| **Clone** | Copiar un repo remoto a tu computadora |
| **Push** | Subir tus cambios al repositorio remoto |
| **Pull** | Bajar los cambios del repositorio remoto |

---

## 🔄 Flujo de trabajo básico

```
Working Directory  →  Staging Area  →  Repositorio local  →  GitHub
   (tus archivos)     (git add)         (git commit)         (git push)
```

## 🛠️ Comandos esenciales

````python
# === INICIAR UN PROYECTO ===

# Inicializar un repositorio nuevo
# git init

# Clonar un repositorio existente de GitHub
# git clone https://github.com/usuario/repositorio.git
````

````python
# === EL CICLO DIARIO ===

# Ver el estado de los archivos
# git status

# Agregar archivos al staging area
# git add archivo.py          # un archivo específico
# git add .                   # todos los archivos modificados

# Hacer un commit (guardar snapshot)
# git commit -m "Descripción clara de qué hiciste"

# Subir cambios a GitHub
# git push origin main
````

````python
# === RAMAS (BRANCHES) ===

# Ver todas las ramas
# git branch

# Crear una rama nueva
# git branch nombre-de-rama

# Cambiar a una rama
# git checkout nombre-de-rama

# Crear y cambiar en un solo paso
# git checkout -b nombre-de-rama

# Fusionar una rama con main
# git checkout main
# git merge nombre-de-rama

# Borrar una rama
# git branch -d nombre-de-rama
````

````python
# === HISTORIAL ===

# Ver el historial de commits
# git log

# Versión compacta
# git log --oneline

# Ver diferencias entre versiones
# git diff
````

````python
# === TRABAJO EN EQUIPO ===

# Bajar los últimos cambios del remoto
# git pull origin main

# Ver repositorios remotos configurados
# git remote -v

# Conectar tu repo local con GitHub
# git remote add origin https://github.com/usuario/repositorio.git
````

## 🗂️ .gitignore

El archivo `.gitignore` le dice a Git qué archivos **no** debe trackear. Muy útil para no subir archivos temporales, credenciales, o carpetas pesadas.

Ejemplo de `.gitignore` para un proyecto de Data Science:

```
# Jupyter checkpoints
.ipynb_checkpoints/

# Entornos virtuales
venv/
__pycache__/

# Variables de entorno (¡nunca subir credenciales!)
.env

# Datasets pesados
*.csv
*.xlsx
```

## 💡 Buenas prácticas

- **Commits frecuentes y descriptivos**: mejor muchos commits chicos que uno enorme
- **Un commit = una idea**: no mezcles cambios no relacionados
- **Siempre `git pull` antes de `git push`** cuando trabajás en equipo
- **Nunca subas credenciales ni contraseñas** al repositorio
- **Usá ramas** para desarrollar features nuevas sin romper lo que ya funciona
